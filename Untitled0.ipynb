```python
import numpy as np

# 1. Criar um array bidimensional
matriz = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print("Matriz original:")
print(matriz)
print("\n")

# 2. Obter dados estatísticos
# a. Média ao longo do eixo 1 (axis=1)
media_axis1 = np.mean(matriz, axis=1)
print("Média ao longo do eixo 1:")
print(media_axis1)
print("\n")

# b. Soma ao longo do eixo 0 (axis=0)
soma_axis0 = np.sum(matriz, axis=0)
print("Soma ao longo do eixo 0:")
print(soma_axis0)
print("\n")

# c. Desvio padrão da matriz completa (sem axis)
desvio_padrao = np.std(matriz)
print("Desvio padrão da matriz completa:")
print(desvio_padrao)
print("\n")

# 3. Obter a transposta da matriz e realizar uma operação com ela
transposta = np.transpose(matriz)
print("Transposta da matriz:")
print(transposta)
print("\n")

# Adicionar 5 a cada elemento da transposta
transposta_adicionada = transposta + 5
print("Transposta adicionada de 5:")
print(transposta_adicionada)
print("\n")

# 4. Realizar um produto escalar entre duas matrizes
matriz_b = np.array([[5, 6, 7], [8, 9, 10], [11, 12, 13]])
produto_escalar = np.dot(matriz, matriz_b)
print("Produto escalar entre duas matrizes:")
print(produto_escalar)
print("\n")

# 5. Criar um filtro para a matriz
filtro = matriz > 45
print("Elementos da matriz que são maiores que 45:")
print(matriz[filtro])
print("\n")

# 6. Realizar alguma operação aritmética
matriz_adicionada = matriz + 3
print("Matriz adicionada de 3:")
print(matriz_adicionada)
print("\n")
```
